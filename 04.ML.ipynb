### 用TSNE將資料降維並可視化出來觀察

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import matplotlib.patheffects as PathEffects

# We import seaborn to make nice plots.
import seaborn as sns
sns.set_style('darkgrid')
sns.set_palette('muted')
sns.set_context("notebook", font_scale=1.5,
                rc={"lines.linewidth": 2.5})

def scatter(x, colors):
    # We choose a color palette with seaborn.
    palette = np.array(sns.color_palette('bright'))

    # We create a scatter plot.
    f = plt.figure(figsize=(7, 7))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=20,
                    c=palette[colors.astype(np.int)])
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    ax.axis('off')
    ax.axis('tight')

    # We add the labels for each digit.
    txts = []
    for i in range(3):
        # Position of each label.
        xtext, ytext = np.median(x[colors == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=30)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=5, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)

    return f, ax, sc, txts

X = features 
y = labels

digits_proj = TSNE(random_state=0).fit_transform(X)
scatter(digits_proj, y)
plt.show()
```

### 使用ML(隨機森林、羅吉斯回歸、支持向量機、單純貝式)來進行分類並建模型

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=0)

#使用隨機森林
RandomForest = RandomForestClassifier(n_estimators=10,max_depth=None,min_samples_leaf=1)
RandomForest.fit(X_train, y_train)
y_pred = RandomForest.predict(X_test)
result_RF = accuracy_score(y_test,y_pred)

#使用羅吉斯回歸
logreg = LogisticRegression(random_state=0)
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
result_LR = accuracy_score(y_test,y_pred)

#使用支持向量機
svc = LinearSVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
result_svc = accuracy_score(y_test,y_pred)

#使用單純貝式
nb = MultinomialNB()
nb.fit(X_train, y_train)
y_pred = nb.predict(X_test)
result_NB = accuracy_score(y_test,y_pred)

print('RandomForestClassifier 預測率：{}'.format(result_RF))
print('LogisticRegression 預測率：{}'.format(result_LR))
print('LinearSVC 預測率：{}'.format(result_svc))
print('MultinomialNB 預測率：{}'.format(result_NB))
```

### 使用訓練好的模型來預測新的留言

```python
import pandas as pd
df13 = pd.read_excel('./comment13.xlsx',engine='openpyxl')
jieba.load_userdict('jieba_data/mydict.txt')

cut_comment = []

for make_list in df13['留言內容']:
    cut_comment.append(make_list)
```

```python
# a = ['加油817是你們堅強的後盾聼送鈡大師','的話817們個個神功護體百毒不侵好啦安啦不要機車哦','這個政府心中只有股票金錢']
aa = tfidf_model.get_feature_names()
b=[]
c=[]
p=0
n=0
agree=[]
for i in cut_comment:
    comment_word = jieba.lcut(i)
#     for text in comment_word:
#     print(comment_word)
    for i in aa:
        if i in comment_word:
            b.append(1)
        else:    
            b.append(0)
    d = np.array(b).reshape(1, -1)
    y_pred = svc.predict(d)
#     agree.append(y_pred)
    if y_pred == 1:
        agree.append(1)
        p += 1
    else:
        agree.append(0)
        n += 1
    b.clear()
print('正面留言:{}\n負面留言:{}'.format(p,n))
```

### 畫出留言的圓餅圖

```python
classified = 'Positive', 'Negative'
counts = [p,n]
colors = ['orange', 'skyblue']
# matplotlib套件顏色列表
# https://matplotlib.org/tutorials/colors/colors.html
explode = (0, 0)   # 切一塊出來距離0.1

plt.pie(counts, explode = explode, labels = classified, colors = colors, autopct = '%1.1f%%', shadow = True, startangle = 140)
plt.axis('equal')
plt.show()
```

### 再加上留言的同意人數進行修正

```python
df13['pred'] = agree
agree_pos = (df13[df13['pred']==1]['同意人數'].sum())/10
agree_neg = (df13[df13['pred']==0]['同意人數'].sum())/10
print('加上同意人數後的正面人數:{}\n加上同意人數後的負面人數:{}'.format(p+agree_pos,n+agree_neg))
```
