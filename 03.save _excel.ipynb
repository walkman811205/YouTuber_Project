```python
from pymongo import MongoClient
# 方法一
# host主機名，27017連線埠
client = MongoClient(host='192.168.157.136', port=27017)
db_auth = client.admin
# 登陸的使用者名稱(username)和密碼(password)
db_auth.authenticate("root", "youtube")
# 連線指定資料庫，資料庫名為db_name
db = client["hello_mongo"]
```

```python
import re
all_comment_data=[]
collection = db.Comment13.find({},{"_id": 0, "user":1,"Comment": 1, "Likes":1})

for comment in collection:
    try:
        User = comment.get('user').split('\n')[1].lstrip()
        Comment = re.sub(r'[^\w]','',comment.get('Comment'))
        Likes = comment.get('Likes').split('\n')[1].lstrip()

        comment_data={

                '留言者':User,
                '留言內容':Comment,
                '同意人數':Likes,
                'label':''
            }
        all_comment_data.append(comment_data)
    except AttributeError as a:
        print(a)
        
```

```python
import pandas as pd
# df = pd.DataFrame(all_job_datas)
df = pd.DataFrame(all_comment_data)
df
```

```python
# df.to_csv('comment12.csv',encoding='utf-8')
df.to_excel('comment13.xlsx', encoding='xlsxwriter')
```
